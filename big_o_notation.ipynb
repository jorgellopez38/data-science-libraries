```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline

#Title your chart "Big O Notation"
#Label your x axis "Elements"
#Label your y axis "Operations"
#Label your curves or make a legend for the curves
#Use LaTex notation where possible
```

```python
#curve "O(1)"
x1 = range(0, 50)
y1 = [0**n + 1 for n in x1]

plt.plot(x1,y1)


```

```python
#label the curve "O(log n)"

x2 = range(0, 50)
y2 = [pow(n, 1) for n in x2]

plt.plot(x2,y2)
plt.yscale('log')
```

```python
# label the curve "O(n)"
x3 = range(0, 50)
y3 = [(n) for n in x3]

plt.plot(x3,y3)
```

```python
# label it "O(n log n)"
x4 = range(0,20)
y4 = [n * pow(n, 1) for n in x4] 

plt.plot(x4,y4)
plt.yscale('log')
```

```python
# label it "O(n^2)"
x5 = range(0,50)
y5 = [n**2 for n in x5] 

plt.plot(x5,y5)
```

```python
# label it "O(2^n)"
x6 = range(0,10)
y6 = [2**n for n in x6] 

plt.plot(x6,y6)
```

```python
# label it "O(n!)"
x7 = range(0,20)
y7 = [np.random.permutation(1) for n in x7]

plt.plot(x7,y7)
```

```python
# label it "O(n^n)"
x8 = range(0,50)
y8 = [n**n for n in x8]  

plt.plot(x8,y8)
```

```python
plt.figure(figsize = (8,6))


plt.plot(x1, y1, c='black', label='$O(1)$', alpha=1.0)
plt.plot(x2, y2, c='purple', label='$O(log n)$', alpha=.8)
plt.plot(x3, y3, c='blue', label='$O(n)$', alpha=.8)
plt.plot(x4, y4, c='green', label='$O(n log n)$', alpha=.8)
plt.plot(x5, y5, c='orange', label='$O(n^2)$', alpha=.8)
plt.plot(x6, y6, c='red', label='$O(2^n)$', alpha=.8)
plt.plot(x7, y7, c='pink', label='$O(n!)$', alpha=.8)
plt.plot(x8, y8, c='brown', label='$O(n^n)$', alpha=.8)


plt.ylim(0, 100)
plt.xlim(0, 20)


plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('operations')
plt.legend()
```
