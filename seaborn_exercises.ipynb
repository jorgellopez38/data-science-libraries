```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
#What does the distribution of petal lengths look like?

sns.histplot(iris.petal_length, kde=True)

```

```python
# Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/ as a 
#hint to how we could visually determine if a correlation exists between two numeric columns.


sns.relplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Length')
plt.ylabel('Width')
```

```python
#Would it be reasonable to predict species based on sepal width and sepal length? For this, 
#you'll visualize two numeric columns through the lense of a categorical column.

sns.relplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Length')
plt.ylabel('Width')
```

```python
#Which features would be best used to predict species?


plt.figure(figsize=(12, 10))
plt.suptitle('Iris Boxplots')

plt.subplot(221)
sns.boxplot(data=iris, y ='sepal_length', x ='species')
plt.title('Sepal Length')

plt.subplot(222)
sns.boxplot(data=iris, y ='sepal_width', x ='species')
plt.title('Sepal Width')

plt.subplot(223)
sns.boxplot(data=iris, y ='petal_length', x ='species')
plt.title('Petal Length')

plt.subplot(224)
sns.boxplot(data=iris, y ='petal_width', x ='species')
plt.title('Petal Width')

plt.subplots_adjust(hspace=0.4)
```

```python
#Load the swiss dataset and read it's documentation. The swiss dataset is available from pydatset 
#rather than seaborn. Create visualizations to answer the following questions:

swiss = data('swiss')
data('swiss', show_doc = True)
```

```python
swiss.sort_values('Catholic', ascending =False)
```

```python
#Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
#(Choose a cutoff point for what constitutes catholic)

swiss[swiss.Catholic > 50].sort_values(by='Catholic', ascending=False)
```

```python
swiss['is_catholic'] = swiss.Catholic > 50
swiss
```

```python
#Does whether or not a province is Catholic influence fertility?

plt.figure(figsize=(5, 8))

sns.boxplot(data=swiss, y ='Infant.Mortality', x ='is_catholic')
plt.title('Da Swiss')


plt.subplots_adjust(hspace=0.4)
```
