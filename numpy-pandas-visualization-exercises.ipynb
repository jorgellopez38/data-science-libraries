```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
# MATPLOTLIB EXERCISES
# 1 - You'll need to write the code that generates the x and y points.

#Use matplotlib to plot the following equation:
#y = x2 âˆ’ x + 2
#Add an anotation for the point 0, 0, the origin.

x = range(-100, 100)
y = [(num**2 - num + 2) for num in x]
```

```python
plt.plot(x, y)
plt.title('$y = x^2 - x + 2')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x, y)
plt.annotate('Origin', xy=(0, 20), xytext=(0, 2500),
             arrowprops={'facecolor': 'blue'})
```

```python
#Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y = âˆšx
# y = x3
# y = 2x
# y = 1 / (x+1)
# You can use functions from the math module to help implement some of the equations above.
```

```python
x1 = range(0, 100)
y1 = [math.sqrt(n) for n in x1]

plt.plot(x1, y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
```

```python
x2 = range(0,20)
y2 = [n**3 for n in x2]


plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
```

```python
x3 = [x * 0.1 for range(0, 150)]
y3 = [n**2 for n in x3]

plt.plot(x3,y3)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')

```

```python
x4 = range(0, 150)
y4 = [1 / (n + 1) for n in x4]

plt.plot(x4,y4)

plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')
```

```python
#Combine the figures you created in the last step into one large figure with 4 subplots.

plt.figure(figsize = (10,6))

#subplot 1
plt.subplot(2,2,1)

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)


#subplot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)



#subplot 3
plt.subplot(223)

plt.plot(x3,y3)

plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x4,y4)

plt.title(r'$ y = \frac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel(r'$\frac{1}{x+1}$')
plt.xlim(0,15)
plt.ylim(0)

plt.tight_layout()
```

```python
#Combine the figures you created in the last step into one figure where each of the 4 equations 
#has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize = (8,6))


plt.plot(x1, y1, c='black', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x2, y2, c='purple', label='$x^{3}$', alpha=.8)
plt.plot(x3, y3, c='blue', label='$2^{x}$', alpha=.8)
plt.plot(x4, y4, c='green', label='$1/(x+1)$', alpha=.8)


plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('My Functions')
plt.legend()

```
